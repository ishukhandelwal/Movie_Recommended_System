```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

```python
movies_data= pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
feature_selected= ['genres','keywords','original_title','cast','director']
print(feature_selected)
```

```python
movies_data[feature_selected] = movies_data[feature_selected].fillna('')
```

```python
combined_features =  movies_data['genres']+" " + movies_data['keywords']+" "+ movies_data['original_title']+ movies_data['cast']+" "+ movies_data['director']
```

```python
print(combined_features)
```

```python
vectorizer= TfidfVectorizer()
```

```python
feature_vector=vectorizer.fit_transform(combined_features)
print(feature_vector)
```

```python
similarity = cosine_similarity(feature_vector)
print(similarity)
```

```python
movie_name= input("enter your favourite movie name")
```

```python
list_of_all_title= movies_data['title'].tolist()
print(list_of_all_title)
```

```python
find_close_match= difflib.get_close_matches(movie_name, list_of_all_title)
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
index_of_the_movie= movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)                             
```

```python
similarity_score= list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
sorted_similiar_movie= sorted(similarity_score, key= lambda x: x[1],reverse=True)
print(sorted_similiar_movie)
```

```python
print("most suggested movie for you \n")
i=1
for movie in sorted_similiar_movie:
    index = movie[0]
    title_from_index= movies_data[movies_data.index== index]['title'].values[0]
    if(i<21):
        print(i,'.', title_from_index)
        i+=1
```
